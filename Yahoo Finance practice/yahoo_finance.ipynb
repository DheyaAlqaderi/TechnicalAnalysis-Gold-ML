```python
from sklearn.preprocessing import MinMaxScaler
# Best for situations where you want to preserve the
# original data's order and relative values.
# it can be sensitive for outliers

df = close.copy()
# Initialize the Min-Max scaler
scaler = MinMaxScaler()
# Fit the scaler on the data and transform it
normalized_data = scaler.fit_transform(df)
# Create a DataFrame with normalized data
normalized_df = pd.DataFrame(normalized_data, columns=df.columns, index=df.index)
normalized_df
```

```python
normalized_df.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize=13)
plt.show()
```

```python
norm.plot(figsize=(15,8), fontsize=15)
plt.legend(fontsize=13)
plt.show()
```

```python
close
```

```python
close.pct_change().dropna(axis=0)
```

```python
ret2 = close.pct_change().dropna()
ret2
```

```python
ret2.describe().T.loc[:, ['mean', 'std']]
```

```python
summary = ret2.describe().T.loc[:, ['mean', 'std']]
summary
```

```python
summary['mean'] = summary['mean']*252# the annual return if we invested 1000$
```

```python
summary['std'] = summary['std']*np.sqrt(252)
```

```python
summary
```

```python
summary.plot.scatter(x='std',y='mean', figsize=(12,8), s=50, fontsize=15)

for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,'std']+0.002, summary.loc[i, 'mean']+0.002), size=15)

plt.xlabel("ann. Risk(std)")
plt.ylabel("ann. Return")
plt.title('Risk/return')
```
