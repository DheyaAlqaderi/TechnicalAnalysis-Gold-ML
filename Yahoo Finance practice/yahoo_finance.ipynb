# Importing and Exporting Stock Price Data from Yahoo Finance

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
ticker = ['AAPL', 'BA', 'KO', 'IBM', 'DIS', 'MSFT']
```

```python
yf.download(ticker, start = '2010-01-01', end = '2023-10-21')
```

```python
stocks = yf.download(ticker, start = '2010-01-01', end = '2023-10-21')
```

```python
stocks.head()
```

```python
stocks.info()
```

```python
stocks.describe(include='all')
```

```python
stocks.to_csv('stocks.csv')
```

```python
stocks = pd.read_csv('stocks.csv', header = [0,1], index_col=[0], parse_dates=[0])
```

```python
stocks.columns
```

```python
stocks.columns = stocks.columns.to_flat_index()
stocks.columns
```

```python
stocks.columns = pd.MultiIndex.from_tuples(stocks.columns)
stocks
```
