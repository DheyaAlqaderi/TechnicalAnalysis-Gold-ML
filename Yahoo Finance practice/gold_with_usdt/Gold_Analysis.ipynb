```python
close = pd.DataFrame(stocks.GLD.Close)
```

```python
close
```

```python
# This normalization is useful when comparing companies within the same industry.
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize=13)
plt.show()
```
