# Simpling Moving Average (SMA) with rolling()

```python
SP500.info()
```

```python
SP500
```

```python
SP500 = SP500.Close.to_frame()
```

```python
SP500
```

```python
SP500.info()
```

```python
SP500.plot(figsize = (12,8), fontsize= 15)
plt.legend(loc = 'upper left', fontsize= 15)
plt.show()
```

```python
SP500 = SP500.loc['2008-12-31':'2018-12-31'].copy()
```

```python
SP500
```

```python
SP500.rolling(window=10)
```

```python
type(SP500.rolling(window=10))
```

```python
SP500.rolling(window=10, min_periods=5).mean()
```

```python
SP500
```

```python
SP500.plot(figsize = (12,8), fontsize= 15)
plt.legend(loc = 'upper left', fontsize= 15)
plt.show()
```

```python
SP500['SMA360'] = SP500.Close.rolling(window=252, min_periods=252).mean()
SP500['SMA30'] = SP500.Close.rolling(window=21, min_periods=21).mean()
SP500['SMA7'] = SP500.Close.rolling(window=5, min_periods=5).mean()
```

```python
SP500
```

```python

SP500.Close.plot(figsize = (12,8), fontsize= 15)
SP500.SMA360.plot(figsize = (12,8), fontsize= 15)
plt.legend(loc = 'upper left', fontsize= 15)
plt.show()
```
